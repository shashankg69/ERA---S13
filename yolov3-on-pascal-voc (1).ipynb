```python
import torch
from lightning_model import Model
from pytorch_lightning import Trainer
from pytorch_lightning.callbacks import ModelSummary, LearningRateMonitor, ModelCheckpoint

model = Model(learning_rate=2*0.00036, num_epochs=40)
trainer = Trainer(max_epochs=40, precision='16-mixed', devices=-1,
                  callbacks=[
                      ModelSummary(max_depth=10),
                      LearningRateMonitor(logging_interval='step'),
                      ModelCheckpoint(
                          dirpath='checkpoints/',
                          filename='{epoch}-{val_loss:.2f}',
                          save_last=True
                      )])
```

```python
trainer.fit(model)
```

```python
torch.save(model.state_dict(), 's13_model.pth')
```

```python
trainer.validate(model)
```
